```python
import cv2 # Mengimpor modul cv2 dari OpenCV
import numpy as np # Mengimpor modul NumPy 
from skimage import data # Mengimpor modul data dari scikit-image
import matplotlib.pyplot as plt # Mengimpor modul pyplot dari matplotlib
%matplotlib inline
```

```python
img = cv2.imread("boneka2.tif", 0) # Membaca gambar 

row, column = img.shape # Mendapatkan dimensi gambar 

img1 = np.zeros((row, column), dtype='uint8') # Membuat array kosong 

# Menentukan rentang nilai minimum dan maksimum
min_range = 10
max_range = 60

# Melakukan perulangan untuk setiap piksel pada gambar
for i in range(row):
    for j in range(column):
        # Jika nilai piksel berada dalam rentang yang ditentukan, set nilai piksel di img1 menjadi 255 (putih)
        if img[i, j] > min_range and img[i, j] < max_range:
            img1[i, j] = 255
        # Jika nilai piksel di luar rentang yang ditentukan, set nilai piksel di img1 menjadi 0 (hitam)
        else:
            img1[i, j] = 0
```

```python
# Membuat subplots dengan ukuran 2x2 untuk menampilkan beberapa gambar dan histogram
fig, axes = plt.subplots(2, 2, figsize=(12, 12))
ax = axes.ravel()

# Menampilkan citra input dan histogramnya pada sumbu pertama dan kedua
ax[0].imshow(img, cmap=plt.cm.gray)
ax[0].set_title("Citra Input")
ax[1].hist(img.ravel(), bins=256)
ax[1].set_title('Histogram Input')

# Menampilkan citra hasil dan histogramnya pada sumbu ketiga dan keempat
ax[2].imshow(img1, cmap=plt.cm.gray)
ax[2].set_title("Citra Output")
ax[3].hist(img1.ravel(), bins=256)
ax[3].set_title('Histogram Output')
```
