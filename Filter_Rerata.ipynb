```python
import matplotlib.pyplot as plt import matplotlib.pyplot as plt # Mengimpor modul pyplot dari matplotlib
%matplotlib inline

# Mengimpor library matplotlib untuk menampilkan gambar dan grafik secara interaktif di notebook
from skimage import data
from skimage.io import imread
from skimage.color import rgb2gray 
import numpy as np # Mengimpor modul NumPy
```

```python
# Membaca gambar
citra1 = imread(fname="goldhill.tif")
citra2 = imread(fname="mobil.tif")

# Mencetak bentuk (shape) dari citra1 dan citra2
print('Shape citra 1 : ', citra1.shape)
print('Shape citra 2 : ', citra2.shape)

# Membuat subplots dengan ukuran 1x2 untuk menampilkan dua gambar
fig, axes = plt.subplots(1, 2, figsize=(10, 10))
ax = axes.ravel()

# Menampilkan citra1 dan citra2 pada sumbu pertama dan kedua dan Mengatur judul untuk setiap citra dengan set_title
ax[0].imshow(citra1, cmap = 'gray')
ax[0].set_title("Citra 1")
ax[1].imshow(citra2, cmap = 'gray')
ax[1].set_title("Citra 2")
```

```python
# Membuat salinan citra dan mengubah tipe datanya menjadi float
copyCitra1 = citra1.copy().astype(float)
copyCitra2 = citra2.copy().astype(float)

m1, n1 = copyCitra1.shape  # Mendapatkan ukuran baris (m1) dan kolom (n1) dari copyCitra1
output1 = np.empty([m1, n1])  # Membuat array kosong 

m2, n2 = copyCitra2.shape  # Mendapatkan ukuran baris (m2) dan kolom (n2) dari copyCitra2
output2 = np.empty([m2, n2])  # Membuat array kosong 

print('Shape copy citra 1 : ', copyCitra1.shape)  # Mencetak bentuk dari copyCitra1
print('Shape output citra 1 : ', output1.shape)  # Mencetak bentuk dari output1

print('m1 : ', m1)  # Mencetak nilai m1 dari copyCitra1
print('n1 : ', n1)  # Mencetak nilai n1 dari copyCitra

print()

print('Shape copy citra 2 : ', copyCitra2.shape)  # Mencetak bentuk (shape) dari copyCitra2
print('Shape output citra 3 : ', output2.shape)  # Mencetak bentuk (shape) dari output2

print('m2 : ', m2)  # Mencetak nilai m2 (jumlah baris) dari copyCitra2.
print('n2 : ', n2)  # Mencetak nilai n2 (jumlah kolom) dari copyCitra2.
print()
```

```python
for baris in range(0, m1-1):
    for kolom in range(0, n1-1):
        a1 = baris  # Menyimpan nilai baris saat ini ke variabel a1.
        b1 = kolom  # Menyimpan nilai kolom saat ini ke variabel b1.

        # Menghitung jumlah dari elemen sekitar pada copyCitra1.
        jumlah = copyCitra1[a1-1, b1-1] + copyCitra1[a1-1, b1] + copyCitra1[a1-1, b1-1] + \
                 copyCitra1[a1, b1-1] + copyCitra1[a1, b1] + copyCitra1[a1, b1+1] + \
                 copyCitra1[a1+1, b1-1] + copyCitra1[a1+1, b1] + copyCitra1[a1+1, b1+1]

        output1[a1, b1] = (1/9 * jumlah)  # Menyimpan nilai rata-rata pada output1.
```

```python
for baris in range(0, m1-1):  # Melakukan iterasi untuk setiap baris dalam rentang 0 hingga m1-1.
    for kolom in range(0, n1-1):  # Melakukan iterasi untuk setiap kolom dalam rentang 0 hingga n1-1.
        a1 = baris  # Menyimpan nilai baris saat ini ke dalam variabel a1.
        b1 = kolom  # Menyimpan nilai kolom saat ini ke dalam variabel b1.

        # Menghitung jumlah dari elemen-elemen sekitar pada copyCitra1.
        jumlah = copyCitra1[a1-1, b1-1] + copyCitra1[a1-1, b1] + copyCitra1[a1-1, b1-1] + \
                 copyCitra1[a1, b1-1] + copyCitra1[a1, b1] + copyCitra1[a1, b1+1] + \
                 copyCitra1[a1+1, b1-1] + copyCitra1[a1+1, b1] + copyCitra1[a1+1, b1+1]

        output1[a1, b1] = (1/9 * jumlah)  # Menyimpan nilai rata-rata pada output1.
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 10))  # Membuat figure dan axes
ax = axes.ravel()  # Meratakan array axes menjadi satu dimensi.

ax[0].imshow(citra1, cmap='gray')  # Menampilkan citra1
ax[0].set_title("Input Citra 1")  # Mengatur judul 

ax[1].imshow(citra2, cmap='gray')  # Menampilkan citra2 
ax[1].set_title("Input Citra 1")  # Mengatur judul 

ax[2].imshow(output1, cmap='gray')  # Menampilkan output1 
ax[2].set_title("Output Citra 1")  # Mengatur judul 

ax[3].imshow(output2, cmap='gray')  # Menampilkan output2 
ax[3].set_title("Output Citra 2")  # Mengatur judul 
```
